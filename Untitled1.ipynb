```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# storing the variable name path
path = (r'C:\Users\kamlesh\Basecamp\Day_5_visualization_for_company_stakeholders\data\data.csv')

# Reading the file and storing in variable called data
read_data = pd.read_csv(path)

#saving the value_count of  loan status
loan_status = read_data['Loan_Status'].value_counts()
```

```python
# Ploting bar graph
plt.bar(loan_status.index, loan_status, color='y', linewidth = 5)
plt.title("loan_status")
plt.xlabel("Yes & No")
plt.ylabel("loan_status_Index")
plt.grid(True)
plt.show()
```

# Using groupby 
groupby is used when we have select from multiple parameters.

```python
property_and_loan = read_data.groupby(['Property_Area','Loan_Status'])
```

```python
property_and_loan = property_and_loan.size().unstack()
```

```python
property_and_loan.plot(kind='bar',stacked=False, color = 'r''b', figsize=(15,10))

plt.title("Property_and_Loan")

plt.xlabel("Property_Area")

plt.ylabel("Loan_Status")

plt.xticks(rotation = 45)

plt.grid()

plt.show()
```

```python
# Expensive education
```

```python
education_and_loan = read_data.groupby(['Education','Loan_Status'])
```

```python
education_and_loan = education_and_loan.size().unstack()
```

```python
#Plotting graph
```

```python
education_and_loan.plot(kind="bar",stacked=True, color= 'y''r', figsize=(8,8))

plt.title("education_and_loan")

plt.xlabel("education")

plt.ylabel("Loan_status")

plt.grid()

plt.xticks(rotation=45)

plt.show()
```

```python
education_and_loan.plot(kind="bar",stacked=False, color= 'y''r', figsize=(8,8))

plt.title("education_and_loan")

plt.xlabel("education")

plt.ylabel("Loan_status")

plt.grid()

plt.xticks(rotation=45)

plt.show()
```

```python
# Subsetting the dataframe using 'Education'Column for graduate
graduate = read_data[read_data['Education']=='Graduate']
```

```python
# subsetting the dataframe using 'Education'colum for not_graduate
not_graduate = read_data[read_data['Education']=='Not Graduate']
```

```python
# plotting the density plot for 'graduate'
graduate['LoanAmount'].plot(kind='density', label='Graduate', color='r')

# Plotting the density plot for 'Not_graduate'
not_graduate['LoanAmount'].plot(kind='density',label='Not Graduate',color='b')

plt.xlabel("Graduate" " " "and" " " "Not Graduate")

plt.grid()

# Plottting automatic legend Disply
plt.legend()
```

### Income  vs Loan

```python
# Creating 3 subplotes
```

```python
fig,(ax_1,ax_2,ax_3)=plt.subplots(1,3, figsize=(20,8))

# Plotting the scatter bar
ax_1.scatter(read_data['ApplicantIncome'],read_data['LoanAmount'],marker="*",s=50)

# Setting axis title 
ax_1.set(title="ApplicantIncome")

# Setting x & y label
ax_1.set(xlabel="ApplicantIncome")

ax_1.set(ylabel="LoanAmount")

plt.grid()

# plotting the scatter bar
ax_2.scatter(read_data['CoapplicantIncome'],read_data['LoanAmount'],color='r')

#Setting axis title
ax_2.set(title='CoapplicantIncome')

# Setting x & y label
ax_2.set(xlabel="CoapplicantIncome")

ax_2.set(ylabel="LoanAmount")

plt.grid()

#Creating sum for totalincome
read_data['TotalIncome'] =read_data['ApplicantIncome']+read_data['CoapplicantIncome']

# Plotting the scatter bar
ax_3.scatter(read_data['TotalIncome'],read_data['LoanAmount'],color='y',marker= "+",alpha=1,s=50)

#Setting axis title
ax_3.set(title='TotalIncome')

# Setting x & y label
ax_3.set(xlabel="TotalIncome")

ax_3.set(ylabel="LoanAmount")
plt.grid()
```

```python
fig,(ax_1)=plt.subplots(1,1, figsize=(10,8))

# Plotting the scatter bar
ax_1.scatter(read_data['ApplicantIncome'],read_data['LoanAmount'],color='orange',edgecolor='red',marker="*",s=100)

# Setting axis title 
ax_1.set(title="ApplicantIncome")

# Setting x & y label
ax_1.set(xlabel="ApplicantIncome")

ax_1.set(ylabel="LoanAmount")

plt.grid()
```

```python
fig,(ax_2)=plt.subplots(1,1, figsize=(10,8))

# plotting the scatter bar
ax_2.scatter(read_data['CoapplicantIncome'],read_data['LoanAmount'],color='r',edgecolor='black',marker="^",alpha=0.75,s=100)
             
#Setting axis title
ax_2.set(title='CoapplicantIncome')

# Setting x & y label
ax_2.set(xlabel="CoapplicantIncome")

ax_2.set(ylabel="LoanAmount")

plt.grid()
```

```python
fig,(ax_3)=plt.subplots(1,1, figsize=(10,8))

#Creating sum for totalincome
read_data['TotalIncome'] =read_data['ApplicantIncome']+read_data['CoapplicantIncome']

# Plotting the scatter bar
ax_3.scatter(read_data['TotalIncome'],read_data['LoanAmount'],color='y',edgecolor='black',marker= "<",alpha=0.75,s=100)

#Setting axis title
ax_3.set(title='TotalIncome')

# Setting x & y label
ax_3.set(xlabel="TotalIncome")

ax_3.set(ylabel="LoanAmount")

plt.grid()

plt.tight_layout()
```
